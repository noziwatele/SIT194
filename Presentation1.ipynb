## 1) $$f(x) = x^2$$

```python
a = 3
b = 7
```

```python
def square(x):
    return x ** 2
```

```python
N_slider = widgets.Text()

def N(x):
    N_slider.value = str(x)

interact(N, x=(1, 100))
```

```python
def integrate(f, a, b, N):
    x = np.linspace(a, b, N, endpoint=False)
    fx = f(x)
    area = np.sum(fx)*(b-a)/N
    return area

ans1 = integrate(square, a, b, int(N_slider.value))
print(ans1)
```

```python
ans2, err2 = quad(square, a, b)
print(ans2)
```

```python
error = (ans2 - ans1) / ans2 * 100
print('% error: ' + str(error))
```

```python
x = np.linspace(0, 10, 20)

x_bar = np.linspace(a, b, int(N_slider.value), endpoint=False)

fig, axes = plt.subplots(2, 1, figsize=(12,12))

#plt.getp(axes)
#plt.tight_layout()

axes[0].plot(x, square(x), 'r')
axes[0].bar(x_bar, square(x_bar), align="edge", width=(x_bar[-1]-x_bar[0]) / len(x_bar), alpha=0.5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Graph of y = x^2')

axes[1].plot(x, square(x), 'r')
section = np.arange(a, b, 1/20.)
axes[1].fill_between(section, square(section), alpha=0.5, edgecolor='0.5')
axes[1].text(0.55 * (a + b), 8, r"$\int_a^b f(x)\mathrm{d}x$",
         horizontalalignment='center', fontsize=20)
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Graph of y = x^2');
```
